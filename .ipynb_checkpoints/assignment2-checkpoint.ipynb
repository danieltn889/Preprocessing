```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import skew, kurtosis


# Load the dataset
df = pd.read_csv("Housing.xls")
print("Dataset shape:", df.shape)
print("\nDataset info:")
print(df.info())
```

```python
# numerical_cols = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']
# categorical_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 
#                    'airconditioning', 'prefarea', 'furnishingstatus']

# Automatically detect numerical and categorical columns

numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()

print(" Numerical columns detected:")
print(numerical_cols)

print("\n Categorical columns detected:")
print(categorical_cols)
```

```python

#1a.
# Select numerical columns automatically
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Custom percentiles: 10%, 25%, 50%, 75%, 90%
percentiles = [0.1, 0.25, 0.5, 0.75, 0.9]

# Generate summary with custom percentiles
numerical_summary = df[numerical_cols].describe(percentiles=percentiles)

print("Statistical Summary of Numerical Variables (with custom percentiles):")
print("="*80)
print(numerical_summary)

# Additional statistics and interpretation
print("\nDetailed Statistics and Interpretation:")
print("="*80)

for col in numerical_cols:
    mean = df[col].mean()
    median = df[col].median()
    min_val = df[col].min()
    max_val = df[col].max()
    std_dev = df[col].std()
    var = df[col].var()
    skewness = df[col].skew()
    kurt_val = df[col].kurtosis()
    
    print(f"\nColumn: {col}")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Min: {min_val}")
    print(f"Max: {max_val}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"Variance: {var:.2f}")
    print(f"Skewness: {skewness:.2f}")
    print(f"Kurtosis: {kurt_val:.2f}")
    
    # Interpretation of skewness
    if skewness > 0.5:
        skew_desc = "right-skewed (long tail on the right)"
    elif skewness < -0.5:
        skew_desc = "left-skewed (long tail on the left)"
    else:
        skew_desc = "approximately symmetric"
    
    # Interpretation of kurtosis
    if kurt_val > 1:
        kurt_desc = "heavy-tailed (more outliers than normal)"
    elif kurt_val < -1:
        kurt_desc = "light-tailed (fewer outliers than normal)"
    else:
        kurt_desc = "approximately normal tails"
    
    print(f"Interpretation: {col} is {skew_desc} with {kurt_desc}.")

# Optional: highlight extreme percentiles for potential outliers
print("\nPercentiles of Interest (10th and 90th):")
print("="*80)
for col in numerical_cols:
    p10 = df[col].quantile(0.1)
    p90 = df[col].quantile(0.9)
    print(f"{col}: 10th percentile = {p10}, 90th percentile = {p90}")
```

```python
# 2a. Detect missing values
print("Missing Values Before Handling:")
print("="*40)
missing_before = df.isnull().sum()
print(missing_before[missing_before > 0])

# 2b. Apply appropriate imputation techniques
# Create a copy for preprocessing
df_clean = df.copy()

# For area (continuous numerical) - use median due to skewness
area_median = df_clean['area'].median()
df_clean['area'].fillna(area_median, inplace=True)

# For bathrooms (discrete numerical) - use mode
bathrooms_mode = df_clean['bathrooms'].mode()[0]
df_clean['bathrooms'].fillna(bathrooms_mode, inplace=True)

print("\nMissing Values After Handling:")
print("="*40)
missing_after = df_clean.isnull().sum()
print(missing_after[missing_after > 0])
```

```python
# 3a. Check for duplicate observations
duplicates = df_clean.duplicated().sum()
print(f"Number of duplicate records: {duplicates}")

if duplicates > 0:
    print("\nDuplicate records found:")
    print(df_clean[df_clean.duplicated()])
    
    # 3b. Remove duplicates
    df_clean = df_clean.drop_duplicates()
    print(f"\nDataset shape after removing duplicates: {df_clean.shape}")
else:
    print("No duplicate records found.")
```

```python
# 4a. Identify inconsistencies
print("Data Types:")
print(df_clean.dtypes)

print("\nCategorical Value Counts:")
for col in categorical_cols:
    print(f"\n{col}:")
    print(df_clean[col].value_counts())

# Check for unrealistic values in numerical columns
print("\nChecking for Unrealistic Values:")
for col in numerical_cols:
    print(f"{col}: Min={df_clean[col].min()}, Max={df_clean[col].max()}")

# 4b. Clean and correct inconsistent data
# Standardize categorical values (ensure consistent casing)
for col in categorical_cols:
    df_clean[col] = df_clean[col].str.lower().str.strip()

# Verify corrections
print("\nCategorical Values After Standardization:")
for col in categorical_cols:
    print(f"\n{col}:")
    print(df_clean[col].value_counts())
```

```python
# 5a. Use IQR method for outlier detection
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

print("Outlier Detection using IQR Method:")
print("="*40)
outlier_info = {}

for col in numerical_cols:
    outliers, lower, upper = detect_outliers_iqr(df_clean, col)
    outlier_info[col] = {
        'count': len(outliers),
        'lower_bound': lower,
        'upper_bound': upper,
        'percentage': (len(outliers) / len(df_clean)) * 100
    }
    print(f"\n{col}:")
    print(f"Outliers: {len(outliers)} ({outlier_info[col]['percentage']:.2f}%)")
    print(f"Bounds: [{lower:.2f}, {upper:.2f}]")

# Visualize outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df_clean[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()

# 5b. Handle outliers - using winsorization for extreme values
df_final = df_clean.copy()

for col in ['price', 'area']:  # Only treat outliers in price and area
    lower_limit = outlier_info[col]['lower_bound']
    upper_limit = outlier_info[col]['upper_bound']
    
    # Winsorize extreme outliers
    df_final[col] = np.where(df_final[col] > upper_limit, upper_limit, 
                            np.where(df_final[col] < lower_limit, lower_limit, df_final[col]))
```

```python
# 6a. Identify variables requiring scaling
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

# 6b. Apply appropriate scaling techniques
scaler_standard = StandardScaler()
scaler_minmax = MinMaxScaler()
scaler_robust = RobustScaler()

# Create scaled versions
df_scaled = df_final.copy()

# Standard Scaling for normally distributed-ish data
df_scaled[['bedrooms', 'bathrooms', 'stories', 'parking']] = scaler_standard.fit_transform(
    df_final[['bedrooms', 'bathrooms', 'stories', 'parking']])

# Robust Scaling for skewed data with outliers
df_scaled[['price', 'area']] = scaler_robust.fit_transform(df_final[['price', 'area']])

print("Data after scaling (first 5 rows):")
print(df_scaled[numerical_cols].head())

# Verify scaling
print("\nScaled Data Statistics:")
print(df_scaled[numerical_cols].describe())
```
